```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("vehicles_us.csv")
df.head(10)
```

```python
df.info()
```

```python
# Crear un histograma de la columna 'price'
lower_bound = df['price'].quantile(0.00)
upper_bound = df['price'].quantile(0.97)

df_filtered = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]

fig_hist = px.histogram(df_filtered, x="price", nbins = 400, color_discrete_sequence = ["#2F6ECE"], template="plotly_white", title="Distribución de Precios")
fig_hist.update_layout(xaxis_title="Precio ($)", yaxis_title="Número de Anuncios", bargap=0.01)
fig_hist.show()
```

```python
# Crear un gráfico de dispersión de 'model_year' vs 'price'
df_f = df[df['model_year'] >= 1995]
fig_scatter = px.scatter(df_f, x="model_year", y="price", title="Precio vs. Año del Modelo (1995 - 2020)", labels={"model_year": "Año del Modelo", "price": "Precio ($)"}, template="plotly_white")
fig_scatter.show()

# Gráfico de dispersión de 'model_year' vs 'price'
# fig_scatter = px.scatter(df, x="model_year", y="price", title="Precio vs. Año del Modelo")
#fig_scatter.show()
```

```python
# APLICAR FILTRO DE PRECIO:
# Se crea un nuevo DataFrame solo con precios menores o iguales a 100,000
df_sin_outliers = df_f[df_f['price'] <= 100000]


# Crear el gráfico de cajas con los datos ya filtrados
fig_boxplot_filtrado = px.box(df_sin_outliers, 
                              x="model_year", 
                              y="price", 
                              title="Distribución de Precios por Año del Modelo (hasta $200k)",
                              labels={"model_year": "Año del Modelo", "price": "Precio ($)"},
                              template="plotly_white")

# Mostrar el gráfico
fig_boxplot_filtrado.show()
```
